```python
df.rename(columns={'M-II':'M2'}, inplace=True)
df
```

```python
df['DV'] = pd.to_numeric(df['DV'], errors='coerce')
df['M2'] = pd.to_numeric(df['M2'], errors='coerce')
df['PP'] = pd.to_numeric(df['PP'], errors='coerce')
df['BEEE'] = pd.to_numeric(df['BEEE'], errors='coerce')
df['FL'] = pd.to_numeric(df['FL'], errors='coerce')
df['FIMS'] = pd.to_numeric(df['FIMS'], errors='coerce')
df.fillna(0, inplace=True)
df
```

```python
df['Total'] = df['DV'] + df['M2'] + df['PP'] + df['BEEE'] + df['FL'] + df['FIMS']
df    
```

CREATING PERCENTAGE COLUMN

```python
df['Percentage'] = (df['Total']/120) *100
df
```

```python
df['Percentage'] = df['Percentage'].round().astype(int)
df
```

```python
def assign_grade(percentage):
    if percentage >= 90:
        return 'A'
    elif percentage >= 80:
        return 'B'
    elif percentage >= 70:
        return 'C'
    elif percentage >= 60:
        return 'D'
    else:
        return 'F'

df['Grade'] = df['Percentage'].apply(assign_grade)
df
```

```python

df = pd.read_excel("MIDMARKS.xlsx")
marks = 10
subjects = ['DV', 'PP', 'BEEE', 'FL', 'FIMS']
pass_fail_counts = {}

for subject in subjects:
    if subject in df.columns:
        numeric_column = pd.to_numeric(df[subject], errors='coerce')
        
        pass_count = (numeric_column >= marks).sum()
        fail_count = (numeric_column < marks).sum()
        pass_fail_counts[subject] = {'Pass': pass_count, 'Fail': fail_count}
    else:
        print(f"Subject '{subject}' not found in the dataset.")
results_df = pd.DataFrame(pass_fail_counts).T
results_df.index.name = "Subject"
results_df.columns = ["Pass", "Fail"]
print(results_df)
results_df.to_excel("Pass_Fail_Counts_All_Subjects.xlsx")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel("MIDMARKS.xlsx")

# Convert the 'PP' column to numeric, coercing errors to NaN
df['PP'] = pd.to_numeric(df['PP'], errors='coerce')

# Define a function to classify the marks
def classify_marks(mark):
    if pd.isna(mark):  # Check for NaN values
        return 'Invalid'  # Or any other category for missing values
    elif mark > 18:
        return 'Best'
    elif 15 <= mark <= 18:
        return 'Very Good'
    elif 12 <= mark < 15:
        return 'Good'
    elif 10 <= mark < 12:
        return 'Average'
    else:
        return 'Poor'

# Apply the classification function to the PP column
df['PP_Category'] = df['PP'].apply(classify_marks)

# Count the number of students in each grade
grade_counts = df['PP_Category'].value_counts()

# Display the grade counts
print("Student Counts by Grade:")
print(grade_counts)

# Plot a bar chart for grade counts
grade_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Students by Grade')
plt.xlabel('Grade')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.show()
```

```python
df['DV'] = pd.to_numeric(df['DV'], errors='coerce')
bins = [0, 1, 2, 3, 4, 5]  
labels = ['0-10', '11-13', '14-15', '16-17', '18-20']
df['DV_Range'] = pd.cut(df['DV'], bins=bins, labels=labels, right=True, include_lowest=True)
DV_range_counts = df['DV_Range'].value_counts().sort_index()
print("Counts of students in each PP marks range:")
print(DV_range_counts)
```

 Number of unique students who scored 20 or more in any subject

```python
import pandas as pd

# Load the dataset
df = pd.read_excel("MIDMARKS.xlsx")

# Convert all columns to numeric, coercing errors to NaN
df = df.apply(pd.to_numeric, errors='coerce')

# Filter students who scored 20 or more in any subject
students_with_20_or_more = df[df.apply(lambda row: any(row >= 20), axis=1)]

# Drop duplicates based on rows (ensuring each student appears only once)
unique_students = students_with_20_or_more.drop_duplicates()

# Get the count of unique students
count_unique_students = len(unique_students)

# Display the filtered dataset and count
print("Students who scored 20 or more in any subject (without repetition):")
print(unique_students)
print(f"\nTotal number of unique students who scored 20 or more in any subject: {count_unique_students}")

# Optionally save to an Excel file
unique_students.to_excel("Unique_Students_20_and_above.xlsx", index=False)
```
