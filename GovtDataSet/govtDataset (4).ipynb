## 1. Importing Required Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt


df=pd.read_csv("crime datset.csv")
df
```

## Checking for Missing Values

```python
df.info()

```


## Basic Statistical Summary

```python


high_charge_sheets = df.nlargest(5, "Charge-Sheeting Rate (Col.18/ Col.19) *100")[["Crime Head", "Charge-Sheeting Rate (Col.18/ Col.19) *100"]]
high_charge_sheets
```

## Data Types and Column Information

```python
top_10_crimes = df.nlargest(10, "Cases Reported during the year")[["Crime Head", "Cases Reported during the year"]]
top_10_crimes
```

```python
avg_charge_sheets = df["Charge-Sheeting Rate (Col.18/ Col.19) *100"].mean()
print(f"Average Charge-Sheeting Rate: {avg_charge_sheets:.2f}%")

```

```python
total_disposed_cases = df["Total Cases Disposed Off by Police (Col.7+ Col.8+ Col.15+ Col.18)"].sum()
print(f"Total Cases Disposed Off by Police: {total_disposed_cases}")
```

## Bar Plot for Category-wise Distribution

```python
sb.set(style="whitegrid")
top_10_crimes = df.nlargest(10, "Cases Reported during the year")

plt.figure(figsize=(12, 6))
sb.barplot(x="Cases Reported during the year", y="Crime Head", data=top_10_crimes, hue="Crime Head", palette="coolwarm", legend=False)
plt.title("Top 10 Crimes by Cases Reported", fontsize=14)
plt.xlabel("Number of Cases", fontsize=12)
plt.ylabel("Crime Type", fontsize=12)
plt.show()
```

## Histogram for Value Distribution

```python
plt.figure(figsize=(8, 8))
df.groupby("Crime Head")["Cases Reported during the year"].sum().nlargest(5).plot.pie(
    autopct="%1.1f%%", cmap="Set2", startangle=90)
plt.title("Top 5 Crimes Distribution")
plt.ylabel("")  # Hide y-label
plt.show()
```

## Line Chart for Yearly Trends

```python
plt.figure(figsize=(12, 6))
sb.lineplot(x="Crime Head", y="Cases Pending Investigation at End of the Year (Col.6- Col.9- Col.19- Col.20)", 
             data=top_10_crimes, marker="o", color="b")
plt.xticks(rotation=90)
plt.title("Pending Cases Trend for Top Crimes")
plt.xlabel("Crime Type")
plt.ylabel("Pending Cases")
plt.show()
```

## Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=["number"])

plt.figure(figsize=(12, 6))
sb.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Crime Data")
plt.show()
```

## Boxplot for Outlier Detection

```python
selected_crimes = ["Murder", "Theft", "Kidnapping & Abduction", "Rape", "Robbery"] 
filtered_df = df[df["Crime Head"].isin(selected_crimes)]

plt.figure(figsize=(12, 6))
sb.lineplot(x="Crime Head", y="Cases Reported during the year", data=filtered_df, marker="o", color="b")
plt.xticks(rotation=45)
plt.title("Reported Cases for Selected Crimes")
plt.xlabel("Crime Type")
plt.ylabel("Cases Reported")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sb.scatterplot(x=df["Cases Reported during the year"], 
                y=df["Cases Pending Investigation at End of the Year (Col.6- Col.9- Col.19- Col.20)"], 
                color="red")
plt.title("Reported Cases vs Pending Cases")
plt.xlabel("Cases Reported")
plt.ylabel("Cases Pending")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sb.boxplot(x=df["Cases Reported during the year"], color="green")
plt.title("Boxplot of Cases Reported")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sb.histplot(df["Cases Reported during the year"], bins=20, kde=True, color="blue")
plt.title("Distribution of Cases Reported")
plt.xlabel("Cases Reported")
plt.ylabel("Frequency")
plt.show()
```

```python
import scipy.stats as stats
mean_cases = df["Cases Reported during the year"].mean()
std_cases = df["Cases Reported during the year"].std()

x = np.linspace(mean_cases - 3*std_cases, mean_cases + 3*std_cases, 100)


pdf = stats.norm.pdf(x, mean_cases, std_cases)

import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.plot(x, pdf, label="Normal Distribution")
plt.xlabel("Cases Reported")
plt.ylabel("Probability Density")
plt.title("Normal Distribution of Reported Cases")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df.groupby("Crime Head")["Cases Reported during the year"].sum().nlargest(5).plot.pie(
    autopct="%1.1f%%", cmap="Set2", startangle=90)
plt.title("Top 5 Crimes Distribution")
plt.ylabel("")  # Hide y-label
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sb.countplot(x='Final Report - Cases Ended as FR Non Cognizable', data=df)
plt.title('Distribution of Crime Types')
plt.xticks(rotation=45)
plt.show()
```

```python
df_subset = df.head(30)

# Create the countplot
plt.figure(figsize=(10, 6))
sb.countplot(x='Chargesheets submitted - Cases Chargesheeted Out of Cases from Previous Year', 
             data=df_subset, 
             hue='Chargesheets submitted - Cases Chargesheeted Out of Cases from Previous Year', 
             palette='viridis', 
             legend=False)
plt.title('Distribution of Chargesheets Submitted (First 30 Rows)')
plt.xticks(rotation=45)
plt.show()
```

```python
df_subset = df.head(20)

# Create the countplot
plt.figure(figsize=(10, 6))
sb.countplot(x='Crime Head', data=df_subset)  # Using the subset of the dataframe
plt.title('Distribution of Crime Heads (First 20 Rows)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by the 'year' column and sum the 'crime_rate'
df.groupby('Total Cases for Investigation (Col.3+ Col.4+ Col.5)')['Cases Stayed at Investigation Stage'].sum().plot(kind='area', figsize=(10, 6), color='skyblue', alpha=0.5)
plt.title('Cumulative Crime Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Cumulative Crime Rate')
plt.show()
```

```python
murder_df = df.head(20)

# Remove rows with missing values in the 'Crime Head' column
murder_df_cleaned = murder_df.dropna(subset=['Crime Head'])

# Plot the count distribution for murder cases with palette
plt.figure(figsize=(10, 10))
sb.countplot(x='Crime Head', data=murder_df_cleaned, hue='Crime Head', palette='viridis', legend=False)
plt.title('Distribution of Murder Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter the dataset for theft cases
theft_df = df[df['Crime Head'].str.contains('Theft', case=False, na=False)]

# Get the count of each unique value in the 'Crime Head' column for theft cases
theft_counts = theft_df['Crime Head'].value_counts()

# Plot a pie chart for the theft cases distribution
plt.figure(figsize=(8, 8))
theft_counts.plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=90, legend=False)
plt.title('Distribution of Theft Cases')
plt.ylabel('')  # Hide the y-axis label for a cleaner chart
plt.show()
```

```python
# Subset the first 10 rows of the dataset for the specific column
df_first_10 = df.head(10)

# Plot the line plot for the first 10 rows
plt.figure(figsize=(10, 6))
plt.plot(df_first_10.index, df_first_10['Cases Stayed at Investigation Stage'], marker='o', color='blue', linestyle='-', linewidth=2)

plt.title('Cases Stayed at Investigation Stage (First 10 Rows)')
plt.xlabel('Row Index')
plt.ylabel('Cases Stayed at Investigation Stage')
plt.grid(True)

plt.show()
```

```python
# Subset the first 20 rows of the dataset
df_first_20 = df.head(20)

# Calculate the value counts of the 'Chargesheets submitted - Cases Chargesheeted Out of Cases during the Year' column
chargesheet_counts = df_first_20['Chargesheets submitted - Cases Chargesheeted Out of Cases during the Year'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(chargesheet_counts, labels=chargesheet_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Distribution of Chargesheets Submitted (First 20 Rows)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
df_first_20 = df.head(20)

# Create a histogram for "Cases Reopened for Investigation"
plt.figure(figsize=(10, 6))
sb.histplot(df_first_20['Cases Reopened for Investigation'], bins=10, kde=True, color='blue')  # Adjust color and bins as needed
plt.title('Histogram of Cases Reopened for Investigation (First 20 Rows)')
plt.xlabel('Cases Reopened for Investigation')
plt.ylabel('Frequency')
plt.show()
```

```python
df_last_20 = df.tail(20)

# Plot the countplot with the 'hue' parameter set
plt.figure(figsize=(10, 6))
sb.countplot(x='Final Report - Cases Ended as Final Report False', 
             data=df_last_20, 
             palette='viridis', 
             hue='Final Report - Cases Ended as Final Report False', 
             legend=False)

plt.title('Final Report - Cases Ended as Final Report False (Last 20 Rows)')
plt.xlabel('Case Outcome')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
df_last_20 = df.tail(20)

# Plot the line plot for "Cases Transferred to other State or Agency" for the last 20 rows
plt.figure(figsize=(10, 6))
sb.lineplot(x=df_last_20.index, y='Cases Transferred to other State or Agency', data=df_last_20, marker='o')

plt.title('Cases Transferred to Other State or Agency (Last 20 Rows)')
plt.xlabel('Index')
plt.ylabel('Cases Transferred')
plt.show()
```
